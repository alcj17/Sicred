```python
#!/usr/bin/env python
# coding: utf-8
#============================================================================================================
# NOME DO SCRIPT: carga_cartao_csv.py
# FUNCAO: Importar dados para a tabela cart√£o no postgresql
#============================================================================================================
# HISTORICO
#
#============================================================================================================

import pyspark
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import functions as Func
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Importa").getOrCreate()

arq = spark.read.csv("C:\Projetos\Dados\MOVIMENTO.csv", header=True, inferSchema=True, sep=";")
arq.show()

grava_associado = arq.withColumnRenamed("id_movimento","id_movimento")\
.withColumnRenamed("vlr_transacao","vlr_transacao")\
.withColumnRenamed("des_transacao","des_transacao")\
.withColumnRenamed("data_movimento","data_movimento")\
.withColumnRenamed("id_cartao","id_cartao")



#Iniciando leitura no Postgresql

print("Iniciando a Carga no POstgresql")

# Data in existing table can be append using below

resumo = spark.read.format("jdbc").option("url","jdbc:postgresql://localhost:5432/desafio")\
.option("dbtable","movimento").option("user","postgres")\
.option("password","123")\
.option("driver","org.postgresql.Driver").load()


grava_associado.write.format("jdbc")\
.option("driver","org.postgresql.Driver")\
.option("url","jdbc:postgresql://localhost:5432/desafio")\
.option("user","postgres")\
.option("password","123").mode('append').option("dbtable","movimento").save()


resumo.show()
```
