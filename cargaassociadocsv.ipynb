```python
#!/usr/bin/env python
# coding: utf-8
#============================================================================================================
# NOME DO SCRIPT: cargaassociadoscsv.py
# FUNCAO: Importar dados dos associados na postgresql
#============================================================================================================
# HISTORICO
#
#============================================================================================================

import pyspark
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import functions as Func
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Importa").getOrCreate()

arq = spark.read.csv("C:\Projetos\Dados\ASSOCIADO.csv", header=True, inferSchema=True, sep=";")
arq.show()

grava_associado = arq.withColumnRenamed("ID","ID")\
.withColumnRenamed("NOME","NOME")\
.withColumnRenamed("SOBRENOME","SOBRENOME")\
.withColumnRenamed("IDADE","IDADE")\
.withColumnRenamed("EMAIL","EMAIL")\


#Iniciando leitura no Postgresql

print("Iniciando a Carga no POstgresql")

# Data in existing table can be append using below

#queryTruncateStage = '(DELETE from associado)' se quiser usar o comando deve-se substituir na linha do dbTable

resumo = spark.read.format("jdbc").option("url","jdbc:postgresql://localhost:5432/desafio")\
.option("dbtable","associado").option("user","postgres")\
.option("password","123")\
.option("driver","org.postgresql.Driver").load()


grava_associado.write.format("jdbc")\
.option("driver","org.postgresql.Driver")\
.option("url","jdbc:postgresql://localhost:5432/desafio")\
.option("user","postgres")\
.option("password","123").mode('append').option("dbtable","associado").save()


resumo.show()
```
